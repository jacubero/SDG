```python
import pandas as pd
```

```python
df_codebook21 = pd.read_excel('../input/SDR2021Database.xlsx', sheet_name='Codebook')
df_codebook21
```

```python
df_codebook20 = pd.read_excel('../input/SDR2020Database.xlsx', sheet_name='Codebook',na_values="NA")
df_codebook20
```

```python
df_codebook19 = pd.read_excel('../input/2019GlobalIndexResults.xlsx', sheet_name='Codebook',na_values="NA")
df_codebook19
```

```python
df_indicator21 = df_codebook21['IndCode']
list_indicator21 = df_indicator21.values.tolist()
set_indicator21 = set(list_indicator21)

df_indicator20 = df_codebook20['Indicator']
list_indicator20 = df_indicator20.values.tolist()
set_indicator20 = set(list_indicator20)

df_indicator19 = df_codebook19['Code']
list_indicator19 = df_indicator19.values.tolist()
set_indicator19 = set(list_indicator19)
```

```python
set_dif2021 = set_indicator20-set_indicator21
set_dif2021
```

```python
set_dif1921 = set_indicator19-set_indicator21
set_dif1921
```

```python
set_dif =set_dif1921.union(set_dif2021)
set_dif
```

```python
list_dif2021 = list(set_dif2021)
df_codebook20[df_codebook20["Indicator"].isin(list_dif2021)]
```

```python
list_dif1921 = list(set_dif1921)
df_codebook19[df_codebook19["Code"].isin(list_dif1921)]
```

```python
df_indicator21 = df_codebook21[['Indicator', 'IndCode']]
df_indicator20 = df_codebook20[['Label', 'Indicator']][df_codebook20["Indicator"].isin(list_dif2021)]
df_indicator20.rename(columns={'Label': 'Indicator', 'Indicator': 'IndCode'}, inplace=True)
df_indicator19 = df_codebook19[['Indicator','Code']][df_codebook19["Code"].isin(list_dif1921)]
df_indicator19.rename(columns={'Code': 'IndCode'}, inplace=True)
df_indicator=pd.concat([df_indicator21,df_indicator20,df_indicator19])
df_indicator
```

```python
records = df_indicator.to_records(index=False)
dict_indicator = dict(records)
dict_indicator
```

```python
df_indicator.to_csv("../output/codebook.csv",index=False)
```

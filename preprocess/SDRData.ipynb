# Preliminaries

```python
import pandas as pd
```

# Data

## 2021

```python
df_data2021 = pd.read_excel('../input/SDR2021Database.xlsx', sheet_name='SDR2021 Data')
df_data2021
```

```python
df_data2021.replace("↑","On track", inplace=True)
df_data2021.replace("➚","Moderately Improving", inplace=True)
df_data2021.replace("→","Stagnating", inplace=True)
df_data2021.replace("↓","Decreasing", inplace=True)
```

```python
df_data2021.rename(columns={'2021 SDG Index Score': 'IndexScore', '2021 SDG Index Rank': 'IndexRank', 'Spillover Score (0-100)': 'SpilloverScore', 'Population in 2020': 'Population'}, inplace=True)
df_data2021.to_csv('../output/data2021.csv',index=False)
```

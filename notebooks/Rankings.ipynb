# Preliminaries

```
import pandas as pd
```

```
csvs_path = 'https://github.com/jacubero/SDG/raw/main/datasets/'
```

# Codes

```
df_codes = pd.read_csv(csvs_path+'codes.csv', engine='python')
df_codes.rename(columns={'alpha-3': 'Country Code ISO3'}, inplace=True)
df_codes
```

```
def map_figtn(alpha: str) -> str:
   base_path = '/flags/'
   path = base_path + 'tn_' + str.lower(str(alpha)) + '-flag.gif'
   
   return path
```

```
df_codes['image']=df_codes.apply(lambda row: map_figtn(row["alpha-2"]), axis=1)
df_codes
```

## 2021

```
df_data2021 = pd.read_csv(csvs_path+'data2021.csv')
```

```
df_sdg_spill = pd.merge(df_data2021[['Country Code ISO3','Country','2021 SDG Index Score','2021 SDG Index Rank','Spillover Score (0-100)','Regions used for the SDR']].head(193), df_codes[['Country Code ISO3','image']], how="left", on="Country Code ISO3")
df_sdg_spill
```

```
df_sdg_spill.sort_values(by=['2021 SDG Index Rank'], inplace=True)
df_sdg_spill.reset_index(drop=True)
```

```
Lines = [".. list-table:: All regions\n", "   :widths: 25 50 25\n", "   :header-rows: 1\n", "   :stub-columns: 1\n", "\n", "   * - Rank\n", "     - Country\n", "     - Score\n"]

for index, row in df_sdg_spill.iterrows():
  Lines.append("   * - " + "{:.0f}".format(row['2021 SDG Index Rank']) +"\n")
  Lines.append("     - .. figure:: " + row['image'] +"\n")
  Lines.append("          :height: 50px\n")
  Lines.append("          :width: 50px\n")
  Lines.append("       " + str(row['Country']) + "\n")
  Lines.append("     - " + "{:.2f}".format(row['2021 SDG Index Score']) +"\n")

with open('sdgindex2021.rst', 'w') as writefile:
    for line in Lines:
        writefile.write(line)
```

```
df_sdg_spill.sort_values(by=['Spillover Score (0-100)'], ascending=False, inplace=True)
df_sdg_spill.reset_index(drop=True)
```

```
Lines = [".. list-table:: All regions\n", "   :widths: 25 50 25\n", "   :header-rows: 1\n", "   :stub-columns: 1\n", "\n", "   * - Rank\n", "     - Country\n", "     - Score\n"]

i=1
for index, row in df_sdg_spill.iterrows():
  Lines.append("   * - " + str(i) +"\n")
  Lines.append("     - .. figure:: " + row['image'] +"\n")
  Lines.append("          :height: 50px\n")
  Lines.append("          :width: 50px\n")
  Lines.append("       " + str(row['Country']) + "\n")
  Lines.append("     - " + "{:.2f}".format(row['Spillover Score (0-100)']) +"\n")
  i +=1

with open('spillover2021.rst', 'w') as writefile:
    for line in Lines:
        writefile.write(line)
```

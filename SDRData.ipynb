# Preliminaries

```python
import pandas as pd
```

```python
excels_path = 'https://github.com/jacubero/SDG/raw/main/datasets/'
```

# Data

## 2021

```python
df_data2021 = pd.read_excel(excels_path+'SDR2021Database.xlsx', sheet_name='SDR2021 Data')
df_data2021
```

```python
df_data2021.replace("↑","On track", inplace=True)
df_data2021.replace("➚","Moderately Improving", inplace=True)
df_data2021.replace("→","Stagnating", inplace=True)
df_data2021.replace("↓","Decreasing", inplace=True)
```

```python
df_data2021.rename(columns={'2021 SDG Index Score': 'IndexScore', '2021 SDG Index Rank': 'IndexRank', 'Spillover Score (0-100)': 'SpilloverScore'}, inplace=True)
df_data2021.to_csv('data2021.csv',index=False)
```

## 2020

```python
df_data2020 = pd.read_excel(excels_path+'SDR2020Database.xlsx', sheet_name='SDR2020 Data')
df_data2020
```

```python
df_data2020.replace("↑","On track", inplace=True)
df_data2020.replace("➚","Moderately Improving", inplace=True)
df_data2020.replace("→","Stagnating", inplace=True)
df_data2020.replace("↓","Decreasing", inplace=True)
```

```python
df_data2020.rename(columns={'id': 'Country Code ISO3', '2020 SDG Index Score': 'IndexScore', '2020 SDG Index Rank': 'IndexRank', 'Spillover Score (0-100)': 'SpilloverScore'}, inplace=True)
df_data2020.to_csv('data2020.csv',index=False) 
```

## 2019

```python
df_data2019 = pd.read_excel(excels_path+'2019GlobalIndexResults.xlsx', sheet_name='SDR2019 Data', skiprows=[0])
df_data2019
```

```python
df_data2019.replace("↑","On track", inplace=True)
df_data2019.replace("➚","Moderately Improving", inplace=True)
df_data2019.replace("→","Stagnating", inplace=True)
df_data2019.replace("↓","Decreasing", inplace=True)
```

```python
df_data2019.rename(columns={'id': 'Country Code ISO3', '2019 Global Index Score (0-100)': 'IndexScore', '2019 Global Index Rank': 'IndexRank', 'Spillover Score (0-100)': 'SpilloverScore', 'Regions used for the SDG Index & Dashboard': 'Regions used for the SDR'}, inplace=True)
df_data2019.to_csv('data2019.csv',index=False)
```

## 2018

```python
df_data2018 = pd.read_excel(excels_path+'2018GlobalIndexResults.xlsx', sheet_name='Sheet1', skiprows=[0])
df_data2018
```
